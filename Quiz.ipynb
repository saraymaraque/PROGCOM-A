```python
import json

# Clase base abstracta para preguntas
class Pregunta:
    def __init__(self, texto, opciones, respuesta_correcta):
        self.texto = texto
        self.opciones = opciones
        self.__respuesta_correcta = respuesta_correcta  # Encapsulada

    def mostrar(self):
        print(f"\n{self.texto}")
        for clave, valor in self.opciones.items():
            print(f"{clave}) {valor}")

    def verificar(self, respuesta_usuario):
        return respuesta_usuario.lower() == self.__respuesta_correcta.lower()

# Subclase que representa preguntas de opciÃ³n mÃºltiple
class PreguntaOpcionMultiple(Pregunta):
    def __init__(self, texto, opciones, respuesta_correcta):
        super().__init__(texto, opciones, respuesta_correcta)

# Clase que maneja el juego
class Quiz:
    def __init__(self, nombre_jugador):
        self.__nombre_jugador = nombre_jugador
        self.__puntaje = 0
        self.__preguntas = []

    def agregar_pregunta(self, pregunta):
        self.__preguntas.append(pregunta)

    def iniciar(self):
        print(f"\nğŸ§  Â¡Bienvenido al Quiz del Planeta Tierra, {self.__nombre_jugador}!")
        for i, pregunta in enumerate(self.__preguntas, start=1):
            print(f"\nPregunta {i}:")
            pregunta.mostrar()
            respuesta = input("Tu respuesta (a/b/c): ")
            if pregunta.verificar(respuesta):
                print("âœ… Â¡Correcto!")
                self.__puntaje += 1
            else:
                print("âŒ Incorrecto.")
        print(f"\nğŸ‰ Tu puntaje final es: {self.__puntaje}/{len(self.__preguntas)}")

# FunciÃ³n que carga preguntas desde el archivo JSON
def cargar_preguntas_desde_archivo(nombre_archivo):
    preguntas = []
    try:
        with open(nombre_archivo, "r", encoding="utf-8") as archivo:
            data = json.load(archivo)
            for categoria, preguntas_categoria in data.items():
                for pregunta_data in preguntas_categoria:
                    texto = pregunta_data["texto"]
                    opciones = pregunta_data["opciones"]
                    respuesta_correcta = pregunta_data["respuesta_correcta"]
                    pregunta = PreguntaOpcionMultiple(texto, opciones, respuesta_correcta)
                    preguntas.append(pregunta)
    except FileNotFoundError:
        print(f"âŒ Archivo '{nombre_archivo}' no encontrado.")
    return preguntas

# FunciÃ³n principal que ejecuta el quiz
def ejecutar_quiz():
    nombre = input("Introduce tu nombre para comenzar el quiz: ")
    juego = Quiz(nombre)
    preguntas = cargar_preguntas_desde_archivo("preguntas_planeta_tierra.json")
    if preguntas:
        for pregunta in preguntas:
            juego.agregar_pregunta(pregunta)
        juego.iniciar()
    else:
        print("âš ï¸ No se pudieron cargar preguntas.")

# Ejecutar el juego
ejecutar_quiz()
```
